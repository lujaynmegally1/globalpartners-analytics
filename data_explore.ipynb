```python
def summary(df):
    return pd.DataFrame({
        "non_null_count": df.notna().sum(),
        "null_count": df.isna().sum(),
        'Data Type': df.dtypes,
        'Unique Values': df.nunique(),
        'Duplicates': df.duplicated().sum()})
```

```python
# 1. Get unique names for options with 0 price
free_option_names = set(order_item_options[order_item_options['OPTION_PRICE'] == 0]['OPTION_NAME'].unique())

# 2. Get unique names for options with price > 0
paid_option_names = set(order_item_options[order_item_options['OPTION_PRICE'] > 0]['OPTION_NAME'].unique())

# 3. Find the overlap (Items that are sometimes free and sometimes paid)
overlap = free_option_names.intersection(paid_option_names)

# 4. Find unique "Always Free" items (Potential fixed discounts or complimentary items)
always_free = free_option_names - paid_option_names

print(f"Total Unique 'Free' Options: {len(free_option_names)}")
print(f"Total Unique 'Paid' Options: {len(paid_option_names)}")
print(f"Number of overlapping items: {len(overlap)}")

if len(overlap) > 0:
    print("\n--- Overlapping Items (Sometimes Discounted) ---")
    print(list(overlap)[:10]) # Showing first 10

if len(always_free) > 0:
    print("\n--- Always Free Items (Standard Comps) ---")
    print(list(always_free)[:10])
```
